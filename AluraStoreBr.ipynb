#1. Análise do faturamento


```python
print("Faturamento Loja 1:", loja1["Preço"].sum())
print("Faturamento Loja 2:", loja2["Preço"].sum())
print("Faturamento Loja 3:", loja3["Preço"].sum())
print("Faturamento Loja 4:", loja4["Preço"].sum())
```

```python
import matplotlib.pyplot as plt

# Faturamento das lojas
faturamento = [
    loja1["Preço"].sum(),
    loja2["Preço"].sum(),
    loja3["Preço"].sum(),
    loja4["Preço"].sum()
]

lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']

plt.figure(figsize=(8,5))
plt.barh(lojas, faturamento, color='teal')
plt.title('Faturamento Total por Loja')
plt.xlabel('Faturamento (R$)')

# Mostrar os valores no final de cada barra
for i, v in enumerate(faturamento):
    plt.text(v + v*0.01, i, f"R$ {v:,.2f}", va='center')

plt.tight_layout()
plt.show()
```

# 2. Vendas por Categoria


```python
# Contando categorias em cada loja
print("Loja 1:\n", loja1['Categoria do Produto'].value_counts())
print("\nLoja 2:\n", loja2['Categoria do Produto'].value_counts())
print("\nLoja 3:\n", loja3['Categoria do Produto'].value_counts())
print("\nLoja 4:\n", loja4['Categoria do Produto'].value_counts())
```

```python
import matplotlib.pyplot as plt

# Lista com os dataframes e nomes das lojas
lojas = [loja1, loja2, loja3, loja4]
nomes = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']

for i, loja in enumerate(lojas):
    contagem = loja['Categoria do Produto'].value_counts()

    plt.figure(figsize=(6,4))
    contagem.plot(kind='bar', color='skyblue')
    plt.title(f'Categorias de Produtos - {nomes[i]}')
    plt.xlabel('Categoria')
    plt.ylabel('Quantidade')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

# 3. Média de Avaliação das Lojas

```python
# Calculando a média das avaliações
print("Loja 1:", loja1["Avaliação da compra"].mean())
print("Loja 2:", loja2["Avaliação da compra"].mean())
print("Loja 3:", loja3["Avaliação da compra"].mean())
print("Loja 4:", loja4["Avaliação da compra"].mean())
```

```python
import matplotlib.pyplot as plt

media_avaliacoes = [
    loja1["Avaliação da compra"].mean(),
    loja2["Avaliação da compra"].mean(),
    loja3["Avaliação da compra"].mean(),
    loja4["Avaliação da compra"].mean()
]

lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']

plt.figure(figsize=(8,5))
plt.bar(lojas, media_avaliacoes, color='skyblue')
plt.title('Média das Avaliações dos Clientes por Loja')
plt.xlabel('Loja')
plt.ylabel('Avaliação Média')
plt.ylim(0, 5)
for i, v in enumerate(media_avaliacoes):
    plt.text(i, v + 0.1, f"{v:.2f}", ha='center')
plt.show()
```

# 4. Produtos Mais e Menos Vendidos

```python
# Contar vendas de cada produto
contagem = loja1['Produto'].value_counts()

# Mostrar os 5 mais vendidos
print("# Mais vendidos:")
print(contagem.head())

# Mostrar os 5 menos vendidos
print("# Menos vendidos:")
print(contagem.tail())
```

```python
import matplotlib.pyplot as plt

# Contar vendas de cada produto na loja1
contagem = loja1['Produto'].value_counts()

# Pegar top 5 mais vendidos e top 5 menos vendidos
mais_vendidos = contagem.head()
menos_vendidos = contagem.tail()

# Gráfico para os 5 mais vendidos
plt.figure(figsize=(10,4))
plt.bar(mais_vendidos.index, mais_vendidos.values, color='green')
plt.title('Top 5 Produtos Mais Vendidos - Loja 1')
plt.xlabel('Produto')
plt.ylabel('Quantidade Vendida')
plt.show()

# Gráfico para os 5 menos vendidos
plt.figure(figsize=(10,4))
plt.bar(menos_vendidos.index, menos_vendidos.values, color='red')
plt.title('Top 5 Produtos Menos Vendidos - Loja 1')
plt.xlabel('Produto')
plt.ylabel('Quantidade Vendida')
plt.show()
```
